```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('startup_funding.csv')
```

```python
df.head()
```

```python
sorted(df['Startup Name'].unique().tolist())
```

```python
df['Investors Name']=df['Investors Name'].fillna('Undisclosed')
```

```python
df['Investors Name']
```

```python
sorted(df['Investors Name'].astype(str).unique().tolist())
```

```python
df.info()
```

```python
df.drop(columns=['Remarks'],inplace=True)
```

```python
df.head()
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.rename(columns={
    'Date dd/mm/yyyy':'Date',
    'Startup Name':'Startup',
    'Industry Vertical':'Industry',
},inplace=True)
```

```python
df.rename(columns={
    'Amount in USD':'amount'
},inplace=True)
```

```python
df['amount']=df['amount'].fillna('0')
```

```python
df['amount']=df['amount'].str.replace(',','')
```

```python
df.head()
```

```python
df=df[df['amount'].str.isdigit()]
```

```python
df['amount']=df['amount'].astype(float)
```

```python
def to_inr(dollar):
    inr=dollar * 82.5
    return inr/10000000
```

```python
df['amount'].apply(to_inr)
```

```python
df['Date'] = df['Date'].str.replace('05/072018','05/07/2018')
```

```python
df.head()
```

```python
df=df.dropna(subset=['Date','Startup','Industry','City  Location','Investors Name','InvestmentnType','amount'])
```

```python
df.to_csv('Startup_Clean.csv',index=False)
```
